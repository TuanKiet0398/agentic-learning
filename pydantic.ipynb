```python
from pydantic import BaseModel
```

```python
from typing import Optional
class Employee(BaseModel):
    name: str
    id: int
    department: Optional[str] = None
    salary: Optional[float] = None
    is_active: Optional[bool] = True
```

```python
emp1 = Employee(name="Alice", id=101, department="HR")
print(emp1)
```

```python
emp2 = Employee(name="Bob", id=102, department="Finance", is_active=False)
print(emp2)
```

```python
from pydantic import BaseModel
from typing import Optional, List

class Classroom(BaseModel):
    class_name: str
    teacher: str
    students: List[str]
    room_number: Optional[int] = None
    is_active: Optional[bool] = True
```

```python
clasroom = Classroom(
    class_name="Math 101",
    teacher="Mr. Smith",
    students=["Alice", "Bob", "Charlie"],
    room_number=202
)
print(clasroom)
```
